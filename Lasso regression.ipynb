```python
%matplotlib inline
from preamble import *
import mglearn
```

```python
from sklearn.model_selection import train_test_split
x,y = mglearn.datasets.load_extended_boston()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state= 0)
```

```python
import numpy as np
from sklearn.linear_model import Lasso
ls = Lasso().fit(x_train,y_train)

print("Training Accuracy: ",ls.score(x_train,y_train))
print("Testing data Accuracy: ",ls.score(x_test,y_test))
print("Number of features used:",np.sum(ls.coef_ != 0) )
```

```python
ls001 = Lasso(alpha=0.01, max_iter=10000).fit(x_train,y_train)

print("Training Accuracy:{:.2f} ".format(ls001.score(x_train,y_train)))
print("Testing data Accuracy: {:.2f}".format(ls001.score(x_test,y_test)))
print("Number of features used:",np.sum(ls001.coef_ != 0) )
```

```python
ls0001 = Lasso(alpha=0.0001, max_iter=10000).fit(x_train,y_train)

print("Training Accuracy:{:.2f} ".format(ls0001.score(x_train,y_train)))
print("Testing data Accuracy: {:.2f}".format(ls0001.score(x_test,y_test)))
print("Number of features used:",np.sum(ls0001.coef_ != 0) )
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1).fit(x_train,y_train)

print("Training Accuracy:{:.2f} ".format(ridge.score(x_train,y_train)))
print("Testing data Accuracy: {:.2f}".format(ridge.score(x_test,y_test)))
print("Number of features used:",np.sum(ridge.coef_ != 0) )
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(ls.coef_, 's', label="Lasso alpha=1")
plt.plot(ls001.coef_, '^', label="Lasso alpha=0.01")
plt.plot(ls0001.coef_, 'v', label="Lasso alpha=0.001")
plt.plot(ridge.coef_, 'o', label="Ridge alpha=0.1")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient Magnitude")
plt.ylim(-25,25)
plt.legend()
```
